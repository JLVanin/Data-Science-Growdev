####Fórmula de resolução da equação de 2° grau 
####Criar um código para obter a solução da equação 2x² + 5x - 75 = 0

```python
import numpy as np

a = 2
b = 5
c = -75

Delta = b**2 - 4*a*c

if Delta < 0:
  print("não possui raiz real")
else:
  x1 = (-b + np.sqrt(Delta))/(2*a)
  x2 = (-b - np.sqrt(Delta))/(2*a)

  print("x’=", x1)
  print("x’’=", x2)
```

####Gráfico da Função Quadrática y = x²

####Função definida por várias sentenças.
São três leis de formação, ou três regras.

1°) Se x <− 2 então f(x) =− 2

2°) Se x é um valor do intervalo [− 2, 2), então f(x) = x

3°) Se x≥2 então f(x) = 2

Gráfico da função no intervalo [− 4, 4)

```python
x = np.arange(-4,4,0.1) #intervalo [A, B] com 0.1 é o passodef f(x):
n = x.shape[0]
y = np.ones(n)

for i in range(n):
  if x[i]<-2:
    y[i]=-2
  if x[i]>=-2 and x[i]<2:
    y[i]=x[i]
  if x[i]>2:
    y[i]=2

plt.plot(x,y,color='green',linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

####1°) Se x <− 3 então f(x) = x + 12
2°) Se x é um valore do intervalo [− 3, 3), então f(x) = x
2

3°) Se x≥3 então f(x) = x + 6

```python
x = np.arange(-6,6,0.1) #intervalo [A, B] com 0.1 é o passodef f(x):
n = x.shape[0]
y = np.ones(n)

def f(x):
  return x + 12
def g(x):
  return x**2
def h(x):
  return x + 6

for i in range(n):
  if x[i]<-3:
    y[i]=f(x[i])
  if x[i]>=-3 and x[i]<3:
    y[i]=g(x[i])
  if x[i]>=3:
    y[i]=h(x[i])
    
plt.plot(x,y,color='green',linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```
