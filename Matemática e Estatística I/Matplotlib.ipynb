####Usar a biblioteca matplotlib para obter o gráfico da função f: D→R no qual D = {x∈Z | − 2≤x < 3} e f(x) = x² + 1

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-2,3,1) #gera uma sequência de números inteiros

def f(x):
  return x**2 + 1
y = f(x)

plt.plot(x,y,'go',color='green',linewidth=3)
```

####Usar a biblioteca matplotlib para obter o gráfico da função f: D→R no qual D = [− 2, 3) e f(x) = x² + 1

```python
x = np.arange(-2,3,0.1) #0.1 é o passo

def f(x):
  return x**2 + 1
y = f(x)

plt.plot(x,y,color='green',linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
```

#### A função constante é definida por
f: D→R

f(x) = K

Sendo K∈R. Dessa forma para qualquer x∈D a função terá o mesmo valor,
portanto na função constante a imagem é um conjunto unitário.
Exemplo
f: R→R, f(x) = 5.
Para qualquer valor de x, f(x) = 5.

```python
x = np.arange(0,40,0.1) #intervalo [A, B] com 0.1 é o passodef f(x):

n = np.shape(x)
y = 5*np.ones(n)

plt.plot(x,y,color='green',linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
```

####Apresente o gráfico da função f: D→R, tal que D = [− 2, 3] e f(x) = 2x + 5.

```python
x = np.arange(-2,3,0.1) #0.1 é o passo

def f(x):
  return 2*x + 5
y = f(x)

plt.plot(x,y,color='green',linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

####Na função f: D→R, tal que D = [− 4, 5] e f(x) =− 3x + 2.
