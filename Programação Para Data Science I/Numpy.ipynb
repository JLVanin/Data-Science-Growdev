##Masked Arrays

```python
# criação de um array de 6 linhas e 4 colunas, com valores variando de 0 a 9 ([0,9[)
a = np.random.randint(0,10,(6,4))

print(a)

r = ma.masked_greater_equal(a, 3)
r
```

```python
print(a[r.mask])
print(a)
```

```python
a[r.mask] = 20
a
```

##Operações

```python
a = np.array([1,4,7,8])
b = np.array([3,20,43,9])
```

##Reshape and Retype

```python
# podemos reformatar a quantidade de dimensões de um array, trocando de uma dimensão para outra
# ex: um array 5x3 pode se tornar um array 3,5. Um array 6x4 pode se tornar um array 2x12.

old_shape = (6,4)
new_shape = (2,12)

a = np.random.randint(0, 10, old_shape)

print(a)

a = np.reshape(a, new_shape)

print(a)
```

```python
# é possível alterar a tipagem de dados de um array
# muito util para utilizar menos espaço em memória

shape = (6,4)

a = np.random.randint(0, 10, shape)
print(a)
print(a.dtype)
print("{} x {}".format(a.size, a.itemsize))
print("{} bytes".format(a.size * a.itemsize))

a[0,0] = 165536

print(a)

a = a.astype(np.int16)

print(a.dtype)

print("{} x {}".format(a.size, a.itemsize))
print("{} bytes".format(a.size * a.itemsize))

print(a)
```

```python
# carregando a imagem
filename = '5fe7707449f50.jpeg'
img = cv2.imread(filename, 0) / 255.0
#cv2_imshow(img)

img_shape = img.shape
print(img_shape)
```

```python
# criação do filtro
shape = (3,3)

filtro_h = np.full(shape, -1)
filtro_v = np.full(shape, -1)

filtro_h[1, :] = 3
filtro_v[:, 1] = 3

print(filtro_h)
print(filtro_v)
```

```python
metade = 1

filtered_image_shape = (img_shape[0] - 2, img_shape[1] - 2)
filtered_image = np.zeros(filtered_image_shape)

print(filtered_image.shape)

for x in range(metade, img_shape[0] - metade):
    for y in range(metade, img_shape[1] - metade):
        gx = np.sum(img[x - metade : x + metade + 1, y - metade: y + metade + 1] * filtro_h)
        gy = np.sum(img[x - metade : x + metade + 1, y - metade: y + metade + 1] * filtro_v)
        filtered_image[x - metade, y - metade] = (gx + gy)

filtered_image *= 255

cv2_imshow(filtered_image)
```

```python
filtered_image_inverse = 255 - filtered_image

cv2_imshow(filtered_image_inverse)
```

```python
mask_r = ma.masked_greater(filtered_image, 20)

r_image = np.copy(filtered_image)

r_image[mask_r.mask] = 255

rgb_image = np.dstack((filtered_image, filtered_image, r_image))

cv2_imshow(rgb_image)
```
