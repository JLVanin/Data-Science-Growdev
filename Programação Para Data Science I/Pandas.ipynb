### Leitura da base de dados

disponível em: [Kaggle database](https://www.kaggle.com/datasets/gabrielrs3/economy-and-population-of-cities-in-brazil-ibge)

```python
# reading dataset
data = pd.read_excel("Cities_Brazil_IBGE.xlsx")
```

```python
data.head()
```

### Pré-processamento

```python

'''
Função: substitui
Descrição: realiza a troca de um caracter pelo outro em uma série de dados
'''
def substitui(dados, str_in, str_out):
    dados = dados.apply(lambda x: str(x).replace(str_in, str_out))    
    dados = dados.astype(float)
    return dados

data['IDHM'] = substitui(data['IDHM'], ',', '.')                # aplica substituição no IDH por minicipio
data['Pib_2014'] = substitui(data['Pib_2014'], ',', '.')        # aplica substituição no valor do PIB
```

```python
data['IDHM'] # exibe os dados alterados e formatados para float
```

### Processamento

- Agrupa os dados por UF
- Calcula a média de todos os campos numéricos
- Projeta o campo IDHM apenas

```python
ufs = data.groupby('LocalUF')

ufs_means = ufs.mean()

ufs_mean_idhm = ufs_means['IDHM']

ufs_mean_idhm
```

### Exibição

Como estamos trabalhando com dados demográficos vamos exibir os mesmos de forma que fiquem visualmente claros. Para isso vamos abrir uma imagem do Brasil e plotar as informações sobre a imagem.

```python
img = cv2.imread('mapa-capitais-brasil.jpg') # abre a imagem e armazena em img
```

```python
cv2_imshow(img) # exibe a imagem na tela
```

### Localização

Para efeitos de localização dos estados na imagem, um 
arquivo auxiliar é utilizado com as coordenadas de 
pixeis (x e y) de cada estado.

```python
coords = pd.read_csv("coordenadas_estados.txt") # lendo coordenadas de cada estado
```

### Dados IDHM

Para exibir os dados vamos pegar o valor minimo das médias e subtrair de cada valor de IDH e então exibir um circulo em cada estado onde o tamanho representa a quantidade de IDH médio dos municipios daquela UF.

```python
minimum = ufs_mean_idhm.min()                   # menor valor das médias por UF

img_idhm = img.copy()                           # cópia da imagem para podermos plotar informações sobre a imagem sem modificar a original

for index, row in coords.iterrows():            # iteração sobre as coordenadas de cada estado
    uf, x, y = row                              # atribuição multipla da uf, e coordenadas x e y
    idhm = ufs_mean_idhm[index] - minimum       # subtrair linha por linha das médias
    img_idhm = cv2.circle(img_idhm,             # plot dos circulos
                          (x,y), 
                          int(100.0 * idhm), 
                          (255,0,0), 2 ) 

cv2_imshow(img_idhm)                            # exibição da imagem
```

### Dados PIB 2014

Nesse caso vamos projetar os dados das médias do pib de 2014 do País

Vamos pegar os valores maximos e minimos e calcular a variação

fazemos o valor médio do PIB por estado - minimo e o resultado dividido pela variação para normalizar os valores entre 0 e 1

```python
ufs_mean_pib_2014 = ufs_means['Pib_2014']

min_pib = ufs_mean_pib_2014.min()
max_pib = ufs_mean_pib_2014.max()

variacao = max_pib - min_pib

img_pib = img.copy()

for index, row in coords.iterrows():
    uf, x, y = row
    pib = (ufs_mean_pib_2014[index] - minimum) / variacao
    img_pib = cv2.circle(img_pib, (x,y), int(30.0 * pib), (255,0,255), 2 )

cv2_imshow(img_pib)
```

### Dados População 2010

Fazemos o somatórios dos dados agrupados por UF

Dividimos os valores por 20 milhões

Exibimos por estado

```python
ufs_sum = ufs.sum()

print(ufs_sum)

ufs_sum_pop_2010 = ufs_sum['PopCenso 2010'] / 40000000

img_pop = img.copy()

for index, row in coords.iterrows():
    uf, x, y = row
    pop = ufs_sum_pop_2010[index]
    img_pop = cv2.circle(img_pop, (x,y), int(30.0 * pop), (255,0,0), 2 )

cv2_imshow(img_pop)
```
