```
airlines = spark.read. format("csv")\
                     . option("header", "true")\
                     . option("inferSchema" , "true")\
                     . load("/databricks-datasets/asa/airlines/2008.csv" )
```

```
import pyspark.sql.functions as F

df = airlines.select('ActualElapsedTime', 'DepDelay', 'Month')

mean_month = df.groupBy('Month')\
               .agg(F.mean('ActualElapsedTime').alias('mean_ActualElapsedTime'),\
               F.mean('DepDelay').alias('mean_DepDelay'))
```

```
display(mean_month)
```

```
origins = ['IND', 'LAS']

amount_voo = airlines.select('Origin')\
                     .filter(airlines.Origin.isin(origins))\
                     .groupBy('Origin')\
                     .count()
```

```
display(amount_voo)
```
