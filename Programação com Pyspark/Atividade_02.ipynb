#### 1. Utilize o spark para ler a tabela com dados de taxis de New York para o mês de Dezembro de 2019

- Dica: o path da tabela é:

'/databricks-datasets/nyctaxi/tripdata/yellow/yellow_tripdata_2019-12.csv.gz'

```
file_path = '/databricks-datasets/nyctaxi/tripdata/yellow/yellow_tripdata_2019-12.csv.gz'

df = spark.read.format("csv")\
               . option("header", "true")\
               . option("inferSchema" , "true")\
               . load(file_path)
```

```
display(df.limit(10))
```

#### 2. Utilize um comando spark para contar quantas linhas tem esse dataframe

```
df.count()
```

#### 3. Agrupe os dados por número de passageiros e encontre o número de corridas de cada um dos valores.

```
import pyspark.sql.functions as F

df2 = df.groupBy('passenger_count')\
        .agg(F.count('passenger_count'))
```

```
display(df2.limit(10))
```

#### 4. DESAFIO: Tente agrupar por dia (1 a 31 de Desembro) e encontre o número de corridas para cada dia.

```
dec_2019 = df.withColumn('day_of_month',\
            F.dayofmonth(df.tpep_pickup_datetime))
                                    
rider_day = dec_2019.groupBy('day_of_month').count()\
                    .sort('day_of_month')
```

```
display(rider_day)
```
